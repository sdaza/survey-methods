```python
# read the original image 
img = imread('../pics/mona-lisa.jpeg', as_gray=True)
imshow(img)
```

```python
# number of pixels
img.shape
```

```python
# loop over different sample sizes``
images = []
samples = [2, 5, 10, 20, 30, 40]
titles = []
for i in samples:
    temp = downscale(img, i, True)
    images.append(temp)
    titles.append('Sample proportion = ' + str(np.round(temp.shape[0]/img.shape[0], 3)))
```

```python
# plot images
show_image_list(images, list_titles=titles, grid=False, 
    figsize=(10, 8), num_cols = 3, title_fontsize=10)
```
